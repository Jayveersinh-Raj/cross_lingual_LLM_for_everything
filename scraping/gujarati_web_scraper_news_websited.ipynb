```python
import requests
import concurrent.futures
from bs4 import BeautifulSoup

# Set the URL of the website
url = 'https://www.gujaratsamachar.com'

# Set the maximum recursion depth
max_depth = 2

# Initialize the set of visited URLs
visited_urls = set()

# Initialize the set of URLs to visit
urls_to_visit = set([url])

# Initialize the cache of already visited URLs
url_cache = {}

# Define a function to send a GET request to a URL and return the HTML content
def get_html(url):
    # Check the cache for the URL
    if url in url_cache:
        return url_cache[url]

    # Send the GET request and get the response
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser', from_encoding='utf-8')

    # Add the URL and HTML content to the cache
    url_cache[url] = soup

    # Return the HTML content
    return soup

# Define a function to extract all the links on a page and add them to the set of URLs to visit
def extract_links(url):
    # Get the HTML content of the page
    soup = get_html(url)

    # Find all the links on the page
    links = soup.find_all('a')

    # Extract the href attribute from each link and add it to the set of URLs to visit
    for link in links:
        link_url = link.get('href')
        if link_url is None:
            continue

        # Check if the link is a relative URL and resolve it to an absolute URL
        if link_url.startswith('/'):
            link_url = url + link_url

        # Check if the link is an external URL and skip it
        if not link_url.startswith(url):
            continue

        # Add the link URL to the set of URLs to visit
        urls_to_visit.add(link_url)

# Define a function to crawl a URL recursively to a specified depth
def crawl(url, depth):
    # Check if the URL has already been visited or if the maximum recursion depth has been reached
    if url in visited_urls or depth > max_depth:
        return

    # Add the URL to the set of visited URLs
    visited_urls.add(url)

    # Extract all the links on the page
    extract_links(url)

    # Recursively crawl each link with a new thread
    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = []
        for link_url in urls_to_visit:
            futures.append(executor.submit(crawl, link_url, depth + 1))
        concurrent.futures.wait(futures)

# Start the recursive crawling with the initial URL
crawl(url, 0)

# Print the set of visited URLs
visited_urls
```

```python
# Define a function to extract all the Gujarati text from a page
gujarati_text = ''

def extract_gujarati_text(url):
    # Get the HTML content of the page
    soup = get_html(url)

    # Find all the elements with Gujarati text on the page
    gujarati_elements = soup.find_all(lambda tag: tag.name == 'div' and 'gujarati' in tag.get('class', []) and tag.text.strip())

    # Extract the Gujarati text from each element
    for element in gujarati_elements:
        gujarati_text += element.text.strip() + '\n'

    # Return the Gujarati text
    return gujarati_text
```
