```python
import random
import pandas as pd

# list of symbol swaps
old = ["ી", "િ", "ુ", "ૂ", "ે", "ૈ", "ો", "ૌ", "ં", "ઁ", "ૃ", "ૄ", "્"]
new = ["િ", "ી", "ૂ", "ુ", "ૈ", "ે", "ૌ", "ો", "ઁ", "ં","ૄ", "ૃ", ""]

  # List of common words
similar_old = ["ત", "ટ", "ગ", "ઘ","ઇ", "ઈ", "સ", "સ", "સ", "શ", "શ", "શ","ષ", "ષ", "ષ", "શ્ર", "શ્ર"," શ્ર", "ઋ", "ૠ"]
similar_new = ["ટ", "ત", "ઘ", "ગ","ઈ", "ઇ", "શ", "ષ", "શ્ર", "સ", "ષ", "શ્ર", "સ", "શ", "શ્ર", "સ", "શ", "ષ","ૠ", "ઋ"]

def mispell_word(word):
    for char in word:
        if char in old:
            index = old.index(char)
            word = word.replace(char, new[index], 1)
        elif char in similar_old:
            index = similar_old.index(char)
            word = word.replace(char, similar_new[index], 1)
    return word

def mispell_gujarati_sentence(sentence):
    words = sentence.split(' ')
    for i in range(len(words)):
        # Randomly decide whether to introduce a mistake in this word
        if random.random() < 0.2:  # 20% chance for each word
            words[i] = mispell_word(words[i])
    return ' '.join(words)

def create_dataset(sentences, filename):
    data = []
    for sentence in sentences:
        incorrect = mispell_gujarati_sentence(sentence)
        data.append({"Correct": sentence, "Incorrect": incorrect})

    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
```
